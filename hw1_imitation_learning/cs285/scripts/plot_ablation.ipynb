```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Train iteration effect (BC)
xdata = [i*2000 for i in range(1, 10)] # 2000 ~ 18000
mean_return = [935.54, 1868.784, 1871.841, 1903.920, 1989.25, 1993.88708, 2059.0444, 2331.8447, 2422.08447]
std_return = [405.30, 530, 678.078, 260.568, 552.36450, 211.187, 441.299, 450.7428, 238.83]
expert = [4713.65 for _ in range(9)]
plt.plot(xdata, mean_return, label="Eval Mean Return", color='blue')
plt.plot(xdata, expert, label="Human Expert", color='orange')
plt.title("Training iteration ablation study")
plt.xlabel("Training Iteration")
plt.ylabel("Eval Mean Return")
plt.ylim(bottom=500, top=5000)
plt.legend()
plt.show()
```
